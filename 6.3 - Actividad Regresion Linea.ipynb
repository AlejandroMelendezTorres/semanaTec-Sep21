6.  **Regresión lineal múltiple**. Como ahora vamos a incluir más de una variable, el modelo se rescribe a :

$$ \hat{y} = \hat{\beta}_0 + \hat{\beta}_1 \cdot x_1 + \hat{\beta}_2 \cdot x_2 + \hat{\beta}_3\cdot x_3 + \ldots \epsilon $$

Elige  las diferentes combinaciones de variables. Ejemplo: Redes y Youtube vs Ventas 

¿Cuál es el valor de beta_0, betas_1 , beta_2 , el score y el score ajustado de 
**Redes y Youtube vs Ventas**, 
**Redes y TV vs Ventas**, 
**Youtube y TV vs Ventas**, 
**Redes y TV**, 
**Tv y Redes vs Ventas**  y 
**Ventas vs Youtube vs Ventas**  ?



```python
# Vamos a definir una función para calcular esta métrica.
def score_adj(X, y, model):
    return 1 - (1-model.score(X, y))*(len(y)-1)/(len(y)-X.shape[1]-1)
```

```python
rl4 = LinearRegression()
# La variable X la ponemos con doble corchete!
rl4.fit(df[['Redes', 'Youtube']], df['sales'])

print('Redes y Youtube')
print('beta_0: ', rl4.intercept_)
print('betas: ', rl4.coef_)
print('Score: ', rl4.score(df[['Redes', 'Youtube']], df['sales']))
print('Score ajustado:', score_adj(df[['Redes', 'Youtube']], df['sales'], rl4))
```

```python
rl5 = LinearRegression()
# La variable X la ponemos con doble corchete!
rl5.fit(df[['Redes', 'TV']], df['sales'])

print('Redes y TV')
print('beta_0: ', rl5.intercept_)
print('betas: ', rl5.coef_)
print('Score: ', rl5.score(df[['Redes', 'TV']], df['sales']))
print('Score ajustado:', score_adj(df[['Redes', 'TV']], df['sales'], rl5))
```

```python
rl6 = LinearRegression()
# La variable X la ponemos con doble corchete!
rl6.fit(df[['Youtube', 'TV']], df['sales'])

print('Youtube y TV')
print('beta_0: ', rl6.intercept_)
print('betas: ', rl6.coef_)
print('Score: ', rl6.score(df[['Youtube', 'TV']], df['sales']))
print('Score ajustado:', score_adj(df[['Youtube', 'TV']], df['sales'], rl6))
```

```python
rl4 = LinearRegression()
# La variable X la ponemos con doble corchete!
rl4.fit(df[['Redes', 'Youtube','TV']], df['sales'])

print('Redes, Youtube y TV')
print('beta_0: ', rl4.intercept_)
print('betas: ', rl4.coef_)
print('Score: ', rl4.score(df[['Redes', 'Youtube', 'TV']], df['sales']))
print('Score ajustado:', score_adj(df[['Redes', 'Youtube', 'TV']], df['sales'], rl4))
```
